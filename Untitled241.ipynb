```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Conv1D, Flatten, Dropout, concatenate
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# ==============================================================
# üìÇ PATH CONFIGURATION
# ==============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Drone Based Crop Yield Estimation & Disease Mapping System"
dataset_path = os.path.join(base_path, "archive", "Crop Yiled with Soil and Weather.csv")

# Output files (ais_csa_ prefix)
model_path = os.path.join(base_path, "ais_csa_skyharvest_model.h5")
scaler_path = os.path.join(base_path, "ais_csa_skyharvest_scaler.pkl")
yaml_path = os.path.join(base_path, "ais_csa_skyharvest_config.yaml")
json_path = os.path.join(base_path, "ais_csa_skyharvest_prediction.json")
csv_path = os.path.join(base_path, "ais_csa_skyharvest_result.csv")

# ==============================================================
# üß† STEP 1 ‚Äî LOAD DATASET
# ==============================================================
print("\n[INFO] Loading dataset...")
df = pd.read_csv(dataset_path)
df = df.dropna()
print(f"[INFO] Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")

# Assume last column is target (Yield)
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# ==============================================================
# ‚öóÔ∏è STEP 2 ‚Äî SCALING
# ==============================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, scaler_path)
print(f"[INFO] Scaler saved at: {scaler_path}")

# Reshape for CNN-LSTM
X_scaled = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================================================
# ‚öôÔ∏è STEP 3 ‚Äî MODEL DEFINITION
# ==============================================================
def create_model(learning_rate=0.001, filters=64, dropout_rate=0.3):
    inputs = Input(shape=(X_train.shape[1], X_train.shape[2]))

    x = Conv1D(filters=filters, kernel_size=1, activation='relu')(inputs)
    x = Flatten()(x)

    lstm_out = LSTM(64, return_sequences=False)(inputs)

    merged = concatenate([x, lstm_out])
    merged = Dropout(dropout_rate)(merged)
    merged = Dense(64, activation='relu')(merged)
    outputs = Dense(1, activation='linear')(merged)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mse', metrics=['mae'])
    return model

# ==============================================================
# üß¨ STEP 4 ‚Äî HYBRID AIS + CSA OPTIMIZATION (SIMULATED)
# ==============================================================
def hybrid_ais_csa_search():
    print("\n[INFO] Running Hybrid AIS + CSA Optimization...")
    best_params = None
    best_score = float('inf')

    # AIS: exploration stage
    learning_rates = np.random.choice([0.0005, 0.001, 0.002, 0.003], 3, replace=False)
    filters = np.random.choice([32, 64, 128], 3, replace=False)
    dropout_rates = np.random.choice([0.2, 0.3, 0.4], 3, replace=False)

    # CSA: random walk around best (LeÃÅvy flight style search)
    for lr in learning_rates:
        for f in filters:
            for dr in dropout_rates:
                lr_perturbed = float(lr + np.random.uniform(-0.0003, 0.0003))
                f_perturbed = int(max(16, f + np.random.randint(-16, 16)))
                dr_perturbed = float(np.clip(dr + np.random.uniform(-0.05, 0.05), 0.1, 0.5))

                model = create_model(lr_perturbed, f_perturbed, dr_perturbed)
                history = model.fit(
                    X_train, y_train,
                    validation_split=0.2,
                    epochs=15,
                    batch_size=16,
                    verbose=0
                )
                val_loss = min(history.history['val_loss'])
                if val_loss < best_score:
                    best_score = val_loss
                    best_params = {
                        'learning_rate': float(lr_perturbed),
                        'filters': int(f_perturbed),
                        'dropout_rate': float(dr_perturbed)
                    }
                    model.save(model_path)

    print(f"[INFO] Best Parameters (Hybrid AIS+CSA): {best_params}")
    return best_params

best_params = hybrid_ais_csa_search()

# ==============================================================
# üß© STEP 5 ‚Äî FINAL MODEL TRAINING
# ==============================================================
print("\n[INFO] Training Final Model...")
model = create_model(**best_params)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

model.save(model_path)
print(f"[INFO] Model saved at: {model_path}")

# ==============================================================
# üìä STEP 6 ‚Äî EVALUATION
# ==============================================================
y_pred = model.predict(X_test).flatten()
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

metrics = {'MAE': float(mae), 'MSE': float(mse), 'RMSE': float(rmse), 'R2_Score': float(r2)}
print("\nüìä MODEL PERFORMANCE METRICS:")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}")

# ==============================================================
# üíæ STEP 7 ‚Äî SAVE RESULTS (FIXED JSON SERIALIZATION)
# ==============================================================
def make_json_serializable(obj):
    """Convert NumPy types to native Python types recursively."""
    if isinstance(obj, dict):
        return {k: make_json_serializable(v) for k, v in obj.items()}
    elif isinstance(obj, (np.integer, np.int32, np.int64)):
        return int(obj)
    elif isinstance(obj, (np.floating, np.float32, np.float64)):
        return float(obj)
    elif isinstance(obj, (list, tuple, np.ndarray)):
        return [make_json_serializable(i) for i in obj]
    else:
        return obj

serializable_metrics = make_json_serializable(metrics)
serializable_params = make_json_serializable(best_params)

results_df = pd.DataFrame({'Actual_Yield': y_test, 'Predicted_Yield': y_pred})
results_df.to_csv(csv_path, index=False)
print(f"[INFO] Results CSV saved at: {csv_path}")

with open(json_path, "w") as f:
    json.dump(
        {'metrics': serializable_metrics, 'best_params': serializable_params},
        f,
        indent=4
    )
print(f"[INFO] JSON saved at: {json_path}")

with open(yaml_path, "w") as f:
    yaml.dump({'optimizer': 'Hybrid AIS+CSA', 'best_params': serializable_params}, f)
print(f"[INFO] YAML saved at: {yaml_path}")

# ==============================================================
# üìà STEP 8 ‚Äî VISUALIZATIONS
# ==============================================================

# 1Ô∏è‚É£ Accuracy Graph
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Hybrid AIS+CSA SkyHarvest Accuracy Graph")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_csa_skyharvest_accuracy_graph.png"))

# 2Ô∏è‚É£ Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap='YlGnBu', annot=True, fmt=".2f")
plt.title("Hybrid AIS+CSA SkyHarvest Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_csa_skyharvest_heatmap.png"))

# 3Ô∏è‚É£ Comparison (Actual vs Predicted)
plt.figure(figsize=(7,7))
plt.scatter(y_test, y_pred, color='green', alpha=0.6, edgecolors='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title("Hybrid AIS+CSA SkyHarvest Comparison Graph")
plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_csa_skyharvest_comparison_graph.png"))

# 4Ô∏è‚É£ Result (Bar Chart)
plt.figure(figsize=(12,6))
sample_count = min(50, len(y_test))
bar_width = 0.35
indices = np.arange(sample_count)
plt.bar(indices, y_test[:sample_count], bar_width, label='Actual', color='skyblue')
plt.bar(indices + bar_width, y_pred[:sample_count], bar_width, label='Predicted', color='orange')
plt.title("Hybrid AIS+CSA SkyHarvest Result Graph (Top 50)")
plt.xlabel("Sample Index")
plt.ylabel("Yield Value")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_csa_skyharvest_result_graph.png"))

# 5Ô∏è‚É£ Prediction (Line Chart)
plt.figure(figsize=(10,6))
plt.plot(y_pred, color='purple', linewidth=2)
plt.title("Hybrid AIS+CSA SkyHarvest Prediction Graph")
plt.xlabel("Sample Index")
plt.ylabel("Predicted Yield")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_csa_skyharvest_prediction_graph.png"))

# ==============================================================
# üñ•Ô∏è STEP 9 ‚Äî DISPLAY RESULTS ON SCREEN
# ==============================================================
print("\nüñ•Ô∏è DISPLAYING SAMPLE RESULTS (Actual vs Predicted):\n")
print(results_df.head(10).to_string(index=False))

print("\nüìà Model Metrics Summary:")
for key, value in metrics.items():
    print(f"{key:<10} : {value:.4f}")

print("\n‚úÖ HYBRID AIS+CSA SkyHarvest pipeline executed successfully!")
print(f"üìÅ All outputs saved under: {base_path}")
```
