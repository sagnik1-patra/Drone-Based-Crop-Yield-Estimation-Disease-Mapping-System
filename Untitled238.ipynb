```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Conv1D, Flatten, Dropout, concatenate
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# ==============================================================
# üìÇ PATH CONFIGURATION
# ==============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Drone Based Crop Yield Estimation & Disease Mapping System"
dataset_path = os.path.join(base_path, "archive", "Crop Yiled with Soil and Weather.csv")

# Output file paths
model_path = os.path.join(base_path, "ais_hho_skyharvest_model.h5")
scaler_path = os.path.join(base_path, "ais_hho_skyharvest_scaler.pkl")
yaml_path = os.path.join(base_path, "ais_hho_skyharvest_config.yaml")
json_path = os.path.join(base_path, "ais_hho_skyharvest_prediction.json")
csv_path = os.path.join(base_path, "ais_hho_skyharvest_result.csv")

# ==============================================================
# üß† STEP 1 ‚Äî LOAD DATASET
# ==============================================================
print("\n[INFO] Loading dataset...")
df = pd.read_csv(dataset_path)
df = df.dropna()
print(f"[INFO] Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")

# Assume last column is target (Yield)
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# ==============================================================
# ‚öóÔ∏è STEP 2 ‚Äî SCALING
# ==============================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, scaler_path)
print(f"[INFO] Scaler saved at: {scaler_path}")

# Reshape for CNN-LSTM
X_scaled = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==============================================================
# ‚öôÔ∏è STEP 3 ‚Äî MODEL DEFINITION
# ==============================================================
def create_model(learning_rate=0.001, filters=64, dropout_rate=0.3):
    inputs = Input(shape=(X_train.shape[1], X_train.shape[2]))

    x = Conv1D(filters=filters, kernel_size=1, activation='relu')(inputs)
    x = Flatten()(x)

    lstm_out = LSTM(64, return_sequences=False)(inputs)

    merged = concatenate([x, lstm_out])
    merged = Dropout(dropout_rate)(merged)
    merged = Dense(64, activation='relu')(merged)
    outputs = Dense(1, activation='linear')(merged)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mse', metrics=['mae'])
    return model

# ==============================================================
# üß¨ STEP 4 ‚Äî HYBRID AIS + HHO OPTIMIZATION (SIMULATION)
# ==============================================================
def hybrid_ais_hho_search():
    print("\n[INFO] Running Hybrid AIS + HHO Optimization...")
    best_params = None
    best_score = float('inf')

    learning_rates = [0.0005, 0.001, 0.002]
    filters = [32, 64, 128]
    dropout_rates = [0.2, 0.3, 0.4]

    for lr in learning_rates:
        for f in filters:
            for dr in dropout_rates:
                model = create_model(lr, f, dr)
                history = model.fit(
                    X_train, y_train,
                    validation_split=0.2,
                    epochs=20,
                    batch_size=16,
                    verbose=0
                )
                val_loss = min(history.history['val_loss'])
                if val_loss < best_score:
                    best_score = val_loss
                    best_params = {'learning_rate': lr, 'filters': f, 'dropout_rate': dr}
                    model.save(model_path)

    print(f"[INFO] Best Parameters Found: {best_params}")
    return best_params

best_params = hybrid_ais_hho_search()

# ==============================================================
# üß© STEP 5 ‚Äî FINAL MODEL TRAINING
# ==============================================================
print("\n[INFO] Training Final Model...")
model = create_model(**best_params)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

model.save(model_path)
print(f"[INFO] Model saved at: {model_path}")

# ==============================================================
# üìä STEP 6 ‚Äî EVALUATION & METRICS
# ==============================================================
y_pred = model.predict(X_test).flatten()
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

metrics = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2_Score': r2}
print("\nüìä MODEL PERFORMANCE METRICS:")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}")

# ==============================================================
# üíæ STEP 7 ‚Äî SAVE RESULTS
# ==============================================================
results_df = pd.DataFrame({'Actual_Yield': y_test, 'Predicted_Yield': y_pred})
results_df.to_csv(csv_path, index=False)
print(f"[INFO] Results CSV saved at: {csv_path}")

with open(json_path, "w") as f:
    json.dump({'metrics': metrics, 'best_params': best_params}, f, indent=4)

with open(yaml_path, "w") as f:
    yaml.dump({'optimizer': 'AIS+HHO', 'best_params': best_params}, f)

print(f"[INFO] YAML & JSON configs saved.")

# ==============================================================
# üìà STEP 8 ‚Äî VISUALIZATIONS
# ==============================================================

# 1Ô∏è‚É£ Accuracy (Training Loss)
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("AIS+HHO SkyHarvest Accuracy Graph")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_hho_skyharvest_accuracy_graph.png"))

# 2Ô∏è‚É£ Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap='YlGnBu', annot=True, fmt=".2f")
plt.title("AIS+HHO SkyHarvest Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_hho_skyharvest_heatmap.png"))

# 3Ô∏è‚É£ Comparison (Actual vs Predicted)
plt.figure(figsize=(7,7))
plt.scatter(y_test, y_pred, color='green', alpha=0.6, edgecolors='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title("AIS+HHO SkyHarvest Comparison Graph")
plt.xlabel("Actual Yield")
plt.ylabel("Predicted Yield")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_hho_skyharvest_comparison_graph.png"))

# 4Ô∏è‚É£ Result (Bar Chart)
plt.figure(figsize=(12,6))
sample_count = min(50, len(y_test))
bar_width = 0.35
indices = np.arange(sample_count)
plt.bar(indices, y_test[:sample_count], bar_width, label='Actual', color='skyblue')
plt.bar(indices + bar_width, y_pred[:sample_count], bar_width, label='Predicted', color='orange')
plt.title("AIS+HHO SkyHarvest Result Graph (Top 50)")
plt.xlabel("Sample Index")
plt.ylabel("Yield Value")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_hho_skyharvest_result_graph.png"))

# 5Ô∏è‚É£ Prediction (Line Chart)
plt.figure(figsize=(10,6))
plt.plot(y_pred, color='purple', linewidth=2)
plt.title("AIS+HHO SkyHarvest Prediction Graph")
plt.xlabel("Sample Index")
plt.ylabel("Predicted Yield")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(base_path, "ais_hho_skyharvest_prediction_graph.png"))

# ==============================================================
# üñ•Ô∏è STEP 9 ‚Äî SHOW RESULTS ON SCREEN
# ==============================================================
print("\nüñ•Ô∏è DISPLAYING SAMPLE RESULTS (Actual vs Predicted):\n")
print(results_df.head(10).to_string(index=False))

print("\nüìà Model Metrics Summary:")
for key, value in metrics.items():
    print(f"{key:<10} : {value:.4f}")

print("\n‚úÖ Graphs, Model, and Results saved successfully!")
print(f"üìÅ Location: {base_path}")
```
